```python
import pymysql 
import prettytable
#python連線到MySQL

input_AC = input("請輸入資料庫帳號:") #####這裡要補做防呆
input_PW = input("請輸入資料庫密碼:")

connect = pymysql.connect(
host = "teaching-db.bo-yuan.net", 
user = input_AC,
passwd = input_PW,
db = input_AC,
charset = "utf8", #不能有-
port = 3306 , 
)
cur = connect.cursor() #操作變數>開始

for i in range(5):
  # print(i)
  print(f"(0)離開程式\n(1)顯示會員列表\n(2)新增會員資料\n(3)更新會員資料\n(4)刪除會員資料")#歡迎的介面,顯示選單功能
  i = int(input("指令:"))
  if i == 0:
    break #應該要用exit()?

  elif i == 1 or i == 3 or i == 4:#顯示目前的資料表 >>> (3)更新、(4)刪除其實也會用到,所以加入or(就可以不用重複寫),然後在裡面再塞i=3&i=4的判斷式
    cur.execute("SELECT * FROM `member`")
    t1 = prettytable.PrettyTable(["編號","姓名","生日","地址"], encoding="utf8") #記得import
    t1.align="l" #後面沒下[?],就是全部,l是靠左
    MB_List = cur.fetchall() #先存個tuple變數
    for x in MB_List:
      t1.add_row((x[0],x[1],x[2],x[3])) #因為是tuple,所以要用小括號,x[0]是編號,1:姓名,2:生日,3:地址 ###最下面有tuple在prettytable的測試
    print(t1)
    t1.clear()
    
    if i == 3:#修改--如果user選3,除了上面,我要做...
      input_id = input("請選擇你要修改的資料編號:") #只有要選擇!!放篩選的條件
      input_name = input("請輸入會員姓名:")
      input_bir = input("請輸入會員生日:") #####這裡要補做防呆
      input_add = input("請輸入會員地址:")
      cur.execute("UPDATE  `member` SET `name`= %s , `birthday`= %s , `address` = %s WHERE `id`= %s" , [input_name ,input_bir ,input_add, input_id])
      connect.commit() #記得要放r
      continue

    elif i == 4:#刪除--如果user選4,除了上面,我要做...
      input_id = input("請選擇你要刪除的資料編號:") #放篩選的條件
      cur.execute("DELETE FROM `member` WHERE `id`= %s",input_id)
      #DELETE FROM `test` WHERE `name` ='aa' AND (`month` >'5'OR `day` <'7')
      connect.commit() #記得要放r
      continue

    continue

  elif i == 2:#新增
    input_name = input("請輸入會員姓名:")
    input_bir = input("請輸入會員生日:") #####這裡要補做防呆
    input_add = input("請輸入會員地址:")
    cur.execute("INSERT INTO `member` (`name`,`birthday`,`address`)VALUES(%s,%s,%s)",[input_name,input_bir,input_add]) #不放id,要先確認是不是有自動編號,不然要放,不問user應該用lastid??
    connect.commit() #記得要放
    continue

  else: #輸入指令外的東西
    print(f"<<<少年ㄟ,眼睛看卡金ㄟ>>>")

connect.close() #連線關閉
```
